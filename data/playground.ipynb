```python
import pandas as pd
```

```python
df = pd.read_csv('results_raw.csv', index_col=0, low_memory=False)
df
```

```python
print(df.StartGroup.unique())
print(df.Status.unique())
print(df.Sex.unique())
print(df.Group.unique())
print(df.Nation.unique())
sum(df.Time_Finish.isna())

```

```python
df.reset_index(drop=True, inplace=True)
df.Group.mask(df.Group == '–', pd.NA, inplace=True)
df.StartGroup.mask(df.StartGroup == '–', pd.NA, inplace=True)
df.Bib.mask(df.Bib == '–', pd.NA, inplace=True)
df.Status.mask(df.Status == '–', pd.NA, inplace=True)
df.fillna(pd.NA, inplace=True)
for c in df.columns:
    if 'Time_' in c:
        new_col = c[5:]
        df[new_col] = pd.to_timedelta(df[c]).apply(lambda x: x.total_seconds())
cols = ['Year', 'Name', 'Nation', 'Status', 'Sex', 'Bib', 'StartGroup', 'Group',
        'Högsta punkten', 'Smågan', 'Mångsbodarna', 'Risberg', 'Evertsberg', 'Oxberg', 'Hökberg', 'Eldris', 'Finish']
df = df[cols]
df.to_csv('results_clean.csv')
```

```python
df = pd.read_csv('results_clean.csv', index_col=0)
finish = df[(df.Status == 'Finished') & (df.Finish.notna())][['Year', 'Finish']]
finish['norm'] = finish.groupby('Year')['Finish'].transform(lambda x: x / x.min())

years = list(range(1994,2021))
years.extend([2022, 2023, 2024])

for y in years:
    f = finish.norm[finish.Year == y].plot.kde(bw_method=0.2,
        xlim=(0.5, 4), title='years 1994-2020 (black), 2022-2024 (red)', ylabel='Density',
        color=('black' if y < 2021 else 'red'), alpha=0.5)
    f.set_xlabel('Finish time (normalized to winner)')

gp = finish.groupby('Year').norm.quantile([0.01, 0.05, 0.1, 0.25, 0.5, 0.75, 0.9, 0.95, 0.99]).unstack()
gp.loc[years].plot()
```

```python
df2011 = df[df['Year'] > 2011]
group_sizes = df2011[df2011['Sex'] == 'W'].groupby('Year').size()
total_counts = df2011.groupby('Year').size()
normalized_group_sizes = group_sizes / total_counts
data = normalized_group_sizes.reset_index().rename(columns={0: 'W'})
data
```

```python
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted", context="notebook")
ax = sns.barplot(
    x=data.Year,
    y=data.W*100,
    color='darkorchid')
ax.set(ylabel='Percentage of female participants')
# Add text label at year 2021
ax.text(data.index[data.Year==2021].values[0], 1, 'COVID-19: no public race', ha='left', va='bottom', rotation=90)

```
